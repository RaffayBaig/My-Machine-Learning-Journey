```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,roc_auc_score,r2_score,confusion_matrix
```

```python
df=pd.read_csv(r"C:\Users\92331\Desktop\ML\logistic_regression\data.csv")
df.head()
```

<div style="border:2px solid black; padding:15px; background-color:lightyellow; text-align:center; border-radius:5px;">
  <h2 style="color:darkred;">DATA CLEANING</h2>
</div>


```python
#for statsitical data
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#not important
df=df.drop(["Unnamed: 32","id"],axis=1)
```

```python
#to check for unique entries in the columns 
df.nunique()
```

```python
#encoding categorical column to numerical
df["diagnosis"]=df["diagnosis"].map({"M":0,"B":1})
```

```python
df["diagnosis"].value_counts().plot(kind="bar")
```

<div style="border:2px solid black; padding:15px; background-color:lightyellow; text-align:center; border-radius:5px;">
  <h2 style="color:darkred;">MODEL BUILDING</h2>
</div>


```python

X=df.drop(['diagnosis'],axis=1)
y=df['diagnosis']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=4)
```

```python
pipeline=Pipeline([
    ("imputer",SimpleImputer(strategy='median')),
    ("scaler",StandardScaler()),
    ("model",LogisticRegression(max_iter=2000))
])
```

```python
pipeline.fit(X_train,y_train)
y_pred=pipeline.predict(X_test)
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, pipeline.predict_proba(X_test)[:, 1]))
```
